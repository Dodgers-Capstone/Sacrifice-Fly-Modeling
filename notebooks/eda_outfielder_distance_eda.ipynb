```python
import polars as pl
import matplotlib.pyplot as plt
import seaborn as sns

on_base = pl.scan_parquet("../data/throw_home_runner_on_third_wide_sprint_arm.parquet")
on_base = on_base.filter(pl.col("event_type").is_in(["sac_fly", "field_out"]))
on_base = on_base.with_columns(
    pl.col("event_type").replace({"sac_fly": "Successful", "field_out": "Unsuccessful"})
)
```

```python
distance_traveled_pd = on_base.select(["distance_traveled_fielder", "event_type"]).drop_nulls().collect().to_pandas()
distance_traveled_all_pd = on_base.select(["distance_traveled_all_fielders", "event_type"]).drop_nulls().collect().to_pandas()
distance_to_home_pd = on_base.select(["distance_catch_to_home", "event_type"]).drop_nulls().collect().to_pandas()
distance_to_home_diff_pd = on_base.select(["distance_to_home_diff", "event_type"]).drop_nulls().collect().to_pandas()
```

```python
# distance traveled
plt.figure(figsize=(8, 6))
sns.boxplot(x='event_type', y='distance_traveled_fielder', data=distance_traveled_pd)
plt.title('Outfielder Distance Traveled by Play Outcome')
plt.xlabel('Play Outcome')
plt.ylabel('Outfieldeer Distance Traveled (ft)')
plt.show()
```

```python
# distance to home plate
plt.figure(figsize=(8, 6))
sns.boxplot(x='event_type', y='distance_catch_to_home', data=distance_to_home_pd)
plt.title('Distance to Home Plate by Play Outcome')
plt.xlabel('Play Outcome')
plt.ylabel('Distance to Home Plate (ft)')
plt.show()
```

```python
# distance to home plate
plt.figure(figsize=(8, 6))
sns.boxplot(x='event_type', y='distance_to_home_diff', data=distance_to_home_diff_pd)
plt.title('Outfielder Distance to Home Plate from batted to catch by Play Outcome')
plt.xlabel('Play Outcome')
plt.ylabel('Distance to Home Plate Difference (ft)')
plt.show()
```

```python
# distance to home plate
plt.figure(figsize=(8, 6))
sns.boxplot(x='event_type', y='distance_to_home_diff', data=distance_to_home_diff_pd)
plt.title('Outfielder Distance to Home Plate from batted to catch by Play Outcome')
plt.xlabel('Play Outcome')
plt.ylabel('Distance to Home Plate Difference (ft)')
plt.show()
```

```python
# --- Plot for distance_traveled_fielder ---
plt.figure(figsize=(7, 6))
sns.boxplot(x='event_type', y='distance_traveled_fielder', data=distance_traveled_pd, width=0.5)
plt.title('Fielder Distance Traveled', fontsize=16, pad=15)
plt.xlabel('Sac Fly Outcome', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel('Distance Traveled (ft)', fontsize=14)
plt.show()

# --- Plot for distance_traveled_all_fielders ---
plt.figure(figsize=(7, 6))
sns.boxplot(x='event_type', y='distance_traveled_all_fielders', data=distance_traveled_all_pd, width=0.5)
plt.title('All Fielders Distance Traveled', fontsize=16, pad=15)
plt.xlabel('Sac Fly Outcome', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel('All Fielders Distance Traveled (ft)', fontsize=14)
plt.show()

# --- Plot for distance_catch_to_home ---
plt.figure(figsize=(7, 6))
sns.boxplot(x='event_type', y='distance_catch_to_home', data=distance_to_home_pd, width=0.5)
plt.title('Distance From Catch to Home', fontsize=16, pad=15)
plt.xlabel('Sac Fly Outcome', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel('Distance From Catch to Home (ft)', fontsize=14)
plt.show()

# --- Plot for distance_to_home_diff ---
plt.figure(figsize=(7, 6))
sns.boxplot(x='event_type', y='distance_to_home_diff', data=distance_to_home_diff_pd, width=0.5)
plt.title('Difference in Distance to Home', fontsize=16, pad=15)
plt.xlabel('Sac Fly Outcome', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylabel('Distance to Home Difference (ft)', fontsize=14)
plt.show()
```
